```python
import matplotlib.pyplot as plt
import numpy as np
from matrix import Matrix
```

```python
matrix1 = Matrix(3, 3)

plt.matshow(matrix1.matrix)
plt.colorbar()  # Add a colorbar to show the values
plt.show()
```

```python
matrix2 = Matrix(10,10)
plt.imshow(matrix2.matrix, cmap='viridis', aspect='auto')
plt.colorbar(label='Value')
plt.title('Heatmap of a 10x10 Matrix')
plt.xlabel('Column Index')
plt.ylabel('Row Index')
plt.show()
```
